```python
from extract_data import extract_data
import pandas as pd
```

```python
data_dir = extract_data()
datas_dict = {key:pd.read_csv(path) for key,path in data_dir.items()}
```

```python
def fill_with_zero(data):
    for column in data.columns:
        data[column] = data[column].fillna(0)
    return data
```

```python
for key, value in datas_dict.items():
    path = '../new_datasets/' + key + '.csv'
    new_df = fill_with_zero(value)
    new_df.to_csv(path, index=False)
```
